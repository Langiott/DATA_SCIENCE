```python
#Livello Proooo
game1 = "Wii Sports"
game2 = "Super Mario Bros."

attribute_list = ["NA_Sales","EU_Sales","JP_Sales"]

def compare_players(game1, game2, attribute_list):
    
    
    
    subset_players = df[(df["Name"] == game1) | (df["Name"]==game2)]
    
   
    subset_compare_attribute = subset_players[attribute_list]
    
    
    subset_compare_attribute_long = subset_compare_attribute.melt(var_name = "Attribute", 
                                                                  value_name = "Rating").Rating
    
    
    
    game11 = [] 
    for i in range(0,2 * len(attribute_list)): 
        if (i %2 == 0): 
            game11.append(subset_compare_attribute_long[i]) 


    game12 = []
    for i in range(0,2 * len(attribute_list)): 
        if (i %2 == 1): 
            game12.append(subset_compare_attribute_long[i]) 
    
    
    
    game11 = np.concatenate((game11, [game11[0]]))
    game12 = np.concatenate((game12, [game12[0]]))
    attribute_list_new = np.concatenate((attribute_list, [attribute_list[0]]))
    
    
    att_labels = np.linspace(start = 0, stop = 2*np.pi, num = len(game11))
    
    
    plt.figure(figsize = (8, 8))
    plt.subplot(polar = True)
    plt.plot(att_labels, game11)
    plt.plot(att_labels, game12)  
    lines, labels = plt.thetagrids(np.degrees(att_labels), labels = attribute_list_new)
    plt.title("Games Sales", y = 1.1, fontdict = {"fontsize" : 20})
    plt.legend(labels = [game1, game2], loc= (0.95, 0.8));
    
    return

game1 = "Wii Sports"
game2 = "Super Mario Bros."

attribute_list = ["NA_Sales","EU_Sales","JP_Sales"]

compare_players(game1,game2,attribute_list)  
```
